```python
movie_budgets_clean1['production_budget'] = movie_budgets_clean1['production_budget'].str.replace(',', '')
```

```python
def remove_dollarsigncommas(data, column):
    data[column] = data[column].str.replace(',','')
    data[column] = data[column].str.replace('$','')
    return print('all done') ; print(data[column].head())
```

```python
remove_dollarsigncommas(movie_budgets_clean1, 'production_budget')
remove_dollarsigncommas(movie_budgets_clean1, 'domestic_gross')
remove_dollarsigncommas(movie_budgets_clean1, 'worldwide_gross')
```

```python
movie_budgets_clean1[['production_budget', 'domestic_gross', 'worldwide_gross']] = movie_budgets_clean1[['production_budget', 'domestic_gross', 'worldwide_gross']].applymap(lambda x: int(x))
```

```python
movie_budgets_clean1['release_date'] = pd.to_datetime(movie_budgets_clean1['release_date'])
```

```python
###Gonna try to map out the relationship between genre runtime and numvotes
votesandprimarygenre = votesandgenres2.groupby('genre_1').mean()[['runtime_minutes', 'averagerating', 'numvotes']]
```

```python
votesandprimarygenre
```

```python
fig, ax1 = plt.subplots(figsize=(10,10))
ax1.set_title('Tertiary Genre\nvs\nAvg. Number of Expect Votes')
ax1.bar(
    votesandprimarygenre.index,
    votesandprimarygenre['runtime_minutes']
)
ax1.set_ylabel('Average Runtime (min)')
ax1.set_xlabel('Genre')
plt.xticks(rotation=45, ha='right')
ax2 = ax1.twinx()
ax2.bar(
    votesandprimarygenre.index,
    votesandprimarygenre['numvotes'],
    color = 'red'
)
ax2.set_ylabel('Average Number of Votes')
```

```python
runtimeinfo = votesandprimarygenre['runtime_minutes']
numvoteinfo = votesandprimarygenre['numvotes']
indices = range(len(votesandprimarygenre))
genreinfo = votesandprimarygenre.index
# Calculate optimal width
width = np.min(np.diff(indices))/3.

fig, ax1 = plt.subplots(figsize=(30,20))
ax1.bar(indices-width/2.,runtimeinfo,width,color='b', label='Runtime(min)')
ax2 = ax1.twinx()
ax2.bar(indices+width/2.,numvoteinfo,width,color='r', label='Average Votes')
tiks = ax.get_xticks().tolist()
ax1.axes.set_xticklabels(genreinfo)
ax1.set_xlabel('Genre')
ax1.set_ylabel('Average Runtime')
ax2.set_ylabel('Average Number of Votes')
ax1.set_xticks(range(27))
ax1.set_title('Average Votes and Runtime\nby Primary Genre')
ax1.legend(loc=0)
ax2.legend(loc=2)
plt.show()
#Seems to align with what were saying in regard to action films having by far the highest average vote count with an average runtime of 100 minutes
```
