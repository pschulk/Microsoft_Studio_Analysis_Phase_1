```python
movie_budgets_clean1 = movie_budgets.copy(
)
```

```python
movie_budgets_clean1['production_budget'] = movie_budgets_clean1['production_budget'].str.replace(',', '')
```

```python
def remove_dollarsigncommas(data, column):
    data[column] = data[column].str.replace(',','')
    data[column] = data[column].str.replace('$','')
    return print('all done') ; print(data[column].head())
```

```python
remove_dollarsigncommas(movie_budgets_clean1, 'production_budget')
remove_dollarsigncommas(movie_budgets_clean1, 'domestic_gross')
remove_dollarsigncommas(movie_budgets_clean1, 'worldwide_gross')
```

```python
movie_budgets_clean1[['production_budget', 'domestic_gross', 'worldwide_gross']] = movie_budgets_clean1[['production_budget', 'domestic_gross', 'worldwide_gross']].applymap(lambda x: int(x))
```

```python
movie_budgets_clean1['release_date'] = pd.to_datetime(movie_budgets_clean1['release_date'])
```

```python
basics_and_ratings = pd.read_sql("""
SELECT *
FROM movie_basics
LEFT JOIN movie_ratings
    ON movie_basics.movie_id = movie_ratings.movie_id
    """, conn )
```

```python
fig, ax = plt.subplots(figsize= (30, 10))
ax.set_title('Top 10 Genres by Average Votes')
ax.bar( top10genresbyvote.index, top10genresbyvote['numvotes'])
ax.set_ylabel('Votes')
ax.set_xlabel('Genre')
### One thing to notice is that action films make up 6 of the top 10 genres. We should likely make some sort of action film
```
